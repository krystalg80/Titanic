```python
import pandas as pd
import numpy as np

df = pd.read_csv('titanic.csv')

mean_fare = df['Fare'].mean()
median_fare = df['Fare'].median()
mode_fare = df['Fare'].mode()
min_fare = df['Fare'].min()
max_fare = df['Fare'].max()
min_age = df['Age'].min()
max_age = df['Age'].max()
mean_age = df['Age'].mean()
age_var = df['Age'].var()

print("Mean Fare:", mean_fare)
print("Median Fare:", median_fare)
print("Mode Fare:", mode_fare)
print("Minimum Fare:", min_fare)
print("Max Fare:", max_fare)
print("Youngest Age:", min_age)
print("Oldest Age:", max_age)
print("Mean Age:", mean_age)
print("Variance:", age_var)

df.hist(column="Age")
df.hist(column="Fare")
df.describe()


# Now try solving some other questions that goes a bit outside of Mean Median and Mode:

# Who was the oldest passenger aboard the ship? (Hint, .max(column))
# How much did the cheapest ticket cost? (Hint, .min(column))
# What was the range of ticket prices? (Hint max - min)
```
